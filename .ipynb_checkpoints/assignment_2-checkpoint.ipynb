**Answer to ex 5.1.3.1. here**
The Barabási–Albert (BA) model is an algorithm for generating random scale-free, undirected networks. It takes into account that in growing real-world networks nodes prefer to link with more connected nodes (= preferential attachment). This is in contrast to the ER model, where interaction partners of a node are randomly chosen and, hence, in a large random network the degree of most nodes is fairly equal to the average degree. Essentially, BA models create more real-world-like networks by developing through the following two steps:

- Growth: add new nodes with links to already existing nodes.
- Preferential attachment: the probability that a link of the new node connects to an existing node depends on that node's degree.

````python
# [Answer to ex. 5.1.3.2. here]
import networkx as nx
%matplotlib inline
import matplotlib.pylab as plt


BA = nx.barabasi_albert_graph(100,2,seed=117)


# [Answer to ex. 5.1.3.3. here]
nx.draw(BA)
plt.show()
````

````python
# [Answer to ex. 5.1.3.4. here]

degrees = [len(list(BA.neighbors(n))) for n in BA.nodes()]

fig, ax = plt.subplots()
plt.hist(degrees, bins=20, rwidth=0.8,density=True)
ax.set_xticks([d for d in degrees])
ax.set_xticklabels(degrees)
plt.show()
````

**Answer to ex 5.1.3.5. here**
The distribution is heavily left-skewed compared to the ER network's one. The ER network's degree distribution is normal distributed, thus has more higher degrees. Due to preferential attachment, new nodes in the growth process are more likely to connect to the more connected nodes (-> rich-gets-richer phenomenon). So, few nodes gradually turn into hubs while most nodes in the network only have a few links. This explains the skewed degree distribution.

````python
# SETUP CODE
# Download the data from here:
# http://socialnetworks.mpi-sws.org/data-wosn2009.html
# and place it in the working directory.
import pandas as pd

data = pd.read_csv(
    "facebook-wall.txt",
    delimiter="\t",
    header=None,
    names=['user1', 'user2', 'timestamp']
)

data.head(10)


def get_network_slice(t0, t1):
    data_t = data.loc[data.timestamp.between(t0, t1)]
    data_t = data_t.groupby(['user1', 'user2']).size().reset_index(name='weight')
    return nx.from_pandas_edgelist(data_t, 'user1', 'user2', 'weight', create_using=nx.DiGraph)


G_t = get_network_slice(data.timestamp.max() - 86400 * 365, data.timestamp.max())
````

````python
# [Answer to ex. 5.2.2. here]
from collections import Counter
def drop_zeros(a_list):
    return [i for i in a_list if i>0]

in_degrees = drop_zeros([G_t.in_degree(n,weight='weight') for n in G_t.nodes()])
in_degreeCount = Counter(in_degrees)
in_deg, in_cnt = zip(*in_degreeCount.items())

out_degrees = drop_zeros([G_t.out_degree(n,weight='weight') for n in G_t.nodes()])
out_degreeCount = Counter(out_degrees)
out_deg, out_cnt = zip(*out_degreeCount.items())

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.scatter(in_deg,in_cnt,marker="o", facecolors='none', edgecolors='r', label='in-degrees')
ax.scatter(out_deg,out_cnt,marker="o", facecolors='none', edgecolors='b', label='out-degrees')
ax.set_yscale('log')
ax.set_xscale('log')
plt.legend(loc='lower left')
````

**Comments to 5.2.2**
- significant number of highly connected nodes and large difference in node degrees.
- real world networks are usually sparse and clustered with degree distributions following a power law similar to what we found here.

> **Ex. 8.1.3**: Answer question in the function `fraction_triangles` below. Explain how `fraction_triangles` is related to  computing the clustering coefficient (using `nx.average_clustering`).
>
>> *Hint:* The following code does the same thing as `fraction_triangles`, but at a scale where you can understand what's going on. If you have a hard time understanding the code in the function you can try to play around with this simpler example
>>
>> ```python
>> import networkx as nx 
>>
>> A  = np.array(
>>     [[0, 1, 1, 0],
>>      [1, 0, 1, 0],
>>      [1, 1, 0, 1],
>>      [0, 0, 1, 0]]
>> )
>>
>> G = nx.from_numpy_array(A)
>> nx.draw(G,with_labels=True)
>>
>> def nth(A, n):
>>     A_ = A.copy()    
>>     for _ in range(1,n):
>>         A = A.dot(A_)
>>     return A
>>
>> a_t = nth(A,3).diagonal().sum()/6
>> n = len(A[:,0])
>> p_t = binom(n, 3)
>> ```


````python
# [Answer to ex. 8.1.3 here]

def make_net(el_, nodes):
    nx_input = el_, 'u1', 'u2', 'meet_count', nx.Graph()
    g = nx.from_pandas_edgelist(*nx_input)
    g.add_nodes_from(nodes)
    return g


from scipy.special import binom

def fraction_triangles(el_, nodes):
    '''
    Compute fraction of actual triangles out 
    of the potential triangles.
    '''
    
    g = make_net(el_, nodes)
    
    #Q.1: what is `A`?: the adjacency matrix which is symmetric and binary
    #Q.2: what does `A**3` do? compute the number of paths between two nodes
    #Q.3: what is diagonal of A_t? the number of actual paths of length 3, 
    # i.e. triangles, which include the person. these are called cycles
    # because they start and end at the same person
    
    # count actual triangles    
    A = nx.to_scipy_sparse_matrix(g)
    A_t = A**3
    a_t = A_t.diagonal().sum()/6
    
    #Q.4: what does `binom(n,3)` compute? the number of triangles including the person
    
    # count potential triangles
    n = len(g.nodes())
    p_t = binom(n, 3)
        
    return a_t/p_t
````

**Answer to 8.1.3.**
The element in row i, column j of `A**3` counts the number of possible paths from node i to node j with pathlength 3.
Therefore, on the diagonal it counts triangles. However, each single triangle is counted 6 times because there are two possible paths with pathlength 3 for each of the three elements forming each triangle. And we have to divide the sum of the diagonal elements by 6.

`nx.average_clustering` calculates the average clustering coefficient of all nodes in the network. The average clustering coefficient takes the average of all nodes' single clustering coefficient. A node's clustering coefficient computes the fraction of a node's linked nodes which are also linked. So, if all nodes in a given network are connected to all possible linked nodes, the clustering coefficient is 1. Therefore, in a triangle we have a clustering coefficient equal to 1. The clustering coefficient will be 0 if no nodes are connected. 

The function `fraction_triangles` is defined as: `a_t/p_t` = actual triangles/potential triangles. The average clustering coefficient is 1 if actual triangles are equal to potential triangles.
